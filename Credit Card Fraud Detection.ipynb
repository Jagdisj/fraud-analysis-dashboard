```python
import pandas as pd

df = pd.read_csv("creditcard.csv")
print(df.head())
print(df.info())
```

```python
print(df['Class'].value_counts())
```

```python
fraud_percent = df['Class'].mean() * 100
print(f"Fraud %: {fraud_percent:.2f}")
```

```python
df['Hour'] = (df['Time'] // 3600) % 24
```

```python
print(df.groupby('Class')['Amount'].mean())
print(df.groupby('Hour')['Class'].sum())
```

```python
import matplotlib.pyplot as plt

df.groupby('Hour')['Class'].sum().plot(kind='bar')
plt.title("Fraud Count by Hour")
plt.show()
```

```python
df.to_csv("cleaned_fraud_data.csv", index=False)
```
